```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/weather_data.csv')
```

```python
df
```

```python
df.rename(columns={'windspeed':'wind-speed'},inplace=True)
```

```python
df.rename(lambda x:x.replace ("-",'_'),axis='columns',inplace=True)
```

```python
# 1) find dataframe where windspeed>5 and event=snow
df[(df['wind_speed']>5) & (df['event']=='Snow')]
```

```python
# 2) find dataframe where windspeed<5 and event = rain
df[(df['wind_speed']<5) & (df['event']=="Rain")]
```

```python
# 3) qus 1 or qus 2
df[(df['wind_speed']>5) & (df['event']=='Snow') | (df['wind_speed']<5 & (df['event']=='Rain'))]
```

```python
# 4) to find day in which temperature>30 or windspeed<4
df['day'][(df['temperature']>30) | (df['wind_speed']<4)]
```

```python
# 5) to print event and day where temperature <20 or windspeed>5
df['event'][(df['temperature']<20) | (df['wind_speed']>5)]
```

```python
# 6) qus 4 and qus 5
df[['day','event']][(df['temperature']>30) | (df['wind_speed']<4) & (df['temperature']<20) | (df['wind_speed']>5)]
```

```python
# 7) to print temperature and event where temperature >=35 and windspeed>=5
df[['temperature','event']][(df['temperature']>=35) & (df['wind_speed']>=5)]
```

```python
# 8) to print event where temperature <=28 and windspeed <=5
df['event'][(df['temperature']<=28) & (df['wind_speed']<=5)]
```

```python
# 9) qus 7 or qus 8
df[['event','temperature']][(df['temperature']>=35) & (df['wind_speed']>=5) | (df['temperature']<=28) & (df['wind_speed']<=5)]
```

```python
# 10) to print dataframe in whichevent=sunny
df[(df['event']=='Sunny')]
```
